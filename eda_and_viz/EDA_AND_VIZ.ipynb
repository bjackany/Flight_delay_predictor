```python
jan_19_20 = pd.concat([jan_2019, jan_2020], ignore_index= True, sort = False)
```

## JANUARY DATA VIZ 2019 and 2020 DATA SIDE BY SIDE

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=2)
axis = sns.countplot(x=jan_19_20['OP_CARRIER'], data=jan_19_20, order=jan_19_20['OP_CARRIER'].value_counts().iloc[0:10].index, orient="v",
hue= 'YEAR')
axis.set_xticklabels(axis.get_xticklabels(), ha='right')
plt.title('TOTAL NUMBER OF FLIGHTS BY AIRLINE IN JANUARY', fontsize=30)
plt.xlabel('AIRLINE', fontsize=24)
plt.ylabel('NUMBER OF FLIGHTS', fontsize=24)
plt.tight_layout()
plt.show()
```
